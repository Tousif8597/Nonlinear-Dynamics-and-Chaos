## **Simulate the Lorenz Attractor**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits import mplot3d
def lorenz(u, t):
    x, y, z = u
    f1 = sigma*(y - x)
    f2 = r*x - y - x*z
    f3 = x*y - b*z
    return np.array([f1, f2, f3])
sigma = 10
b = 8/3
r = 28
t = np.linspace(0, 100, 10000)
ax = plt.axes(projection = '3d')
for i in range(6):
    u0 = np.random.random(3)
    sol = odeint(lorenz, u0, t)
    X, Y, Z = sol.T
    ax.plot(X, Y, Z, lw = 0.1)
plt.title(r"Lorenz Attractor for $\sigma$ = 10, $\beta$ = $\frac{8}{3}$ and $\rho$ = 28 for different initial conditions")
plt.show()
```

## **Orbital Period of X**

```python
from scipy.signal import find_peaks
```

```python
peaks, _ = find_peaks(X)
```

```python
OP = np.mean(np.diff(t[peaks]))
```

```python
print("Orbital Period =",OP)
```

## **For attractor reconstruction, time delay should of 1/4th of orbital period**

```python
tau = int(OP*100/4)
```

```python
plt.plot(X[:-tau], X[30:])
```
