```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit #For Faster computation
@njit
def Bifurcation_Diagram(a, b, r_points, n_stable, n_points):
    """
    Parameters:
    a = smallest r value
    b = maximum r value
    r_points = Total number of r points in (1, 4] span.
    n_stable = Total no of iterations taken to get equilibrium value of x for a fixed r value.
    n_points = Total no of iterations for a fixed r value.
    """
    R = np.linspace(a, b, r_points)
    R_value = np.zeros((r_points, n_points))
    X = []
    for i in range(r_points):
        r = R[i]
        x = np.random.random()
        for j in range(n_stable):
            x = r*x*(1 - x)
        for k in range(n_points):
            R_value[i, k] = r
            x = r*x*(1 - x)
            X.append(x)
    return R_value, X
```

```python
R_mat, X = Bifurcation_Diagram(2.98, 4, 900, 200, 300)
R = R_mat.reshape(900*300, 1)
plt.figure(figsize = (10, 6))
plt.scatter(R, X, s = 0.001,color = 'crimson')
plt.axis('off')
plt.show()
```
